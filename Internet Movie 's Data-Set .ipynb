# Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
def rating(r):
    if r>=7.0:
        return "Excellent"
    elif r>6.0:
        return "Good"
    else:
        return "Average"
    
```

```python
s["Rating-set"]=s["Rating"].apply(rating)
s
```
